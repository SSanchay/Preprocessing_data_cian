```python
#Создаем новые признаки-столбцы название метро и расстояние пешком от дома до него
metro = cian_df['Метро'].apply(lambda x: str(x).split('('))

#название метро
cian_df['metro_name'] = [metro[i][0][3:].strip() \
    if metro[i][0][0] == 'м' else None \
    for i in range(len(metro))]

#расстояние в минутах от дома до метро
cian_df['metro_radius_foot'] = [int(metro[i][1].split()[0]) \
    if metro[i][0][0] == 'м' and metro[i][1].split()[0] != 'None' else None \
    for i in range(len(metro))]

#Ставим параметр foot_car = 3 для машины и 1 для пешехода
cian_df['metro_radius_foot_car'] = [None if metro[i][0][0] != 'м' else \
    foot_car if len(metro[i][1].split()) == 4 else 1 \
     for i in range(len(metro))]

#расстояние в минутах пешком от дома до метро
cian_df['metro_radius_foot'] = cian_df['metro_radius_foot']*cian_df['metro_radius_foot_car']
cian_df.drop(columns=['metro_radius_foot_car', 'Метро'], inplace = True)
```

```python
### Сортируем по улице и дому и добавляем предыдущие метро и расстояние там, где пусто
cian_df[['city', 'street+house']] = cian_df['Адрес'].str.split(', ', 1,  expand=True)
cian_df[['street', 'house']] = cian_df['street+house'].str.split(', ', 1,  expand=True)

cian_df = cian_df.sort_values(by=['street', 'house'])
cian_df['metro_name'].fillna(method='ffill', inplace=True)
cian_df['metro_radius_foot'].fillna(method='ffill', inplace=True)

cian_df.drop(columns=['city', 'street+house', 'street', 'house'], inplace=True)
```

```python
# Также как перевести метро в числовую фичу (категориальная нужна скорее всего)
cian_df['district_by_metro'] = cian_df['metro_name'].apply(get_key)
cian_df.drop(columns='metro_name', inplace=True)

# Предобработка категориальных фичей, используем one-hot encoding
cian_df = pd.concat((cian_df, pd.get_dummies(cian_df['district_by_metro'], prefix='district')), axis=1)
cian_df.drop(columns='district_by_metro', inplace=True)
```

```python
# Категориальный признак мусоропровод: 1 (Да) и 0 (Нет), nan заменил на 0
cian_df['garbage_chute'] = cian_df['Мусоропровод'].apply(lambda x: 1 if x == 'Да' else 0)
cian_df = cian_df.drop(columns='Мусоропровод')
```

```python
#Категориальный признак лифт
elevator = cian_df['Лифт'].str.split(',').fillna(0)

cian_df['elevator_pass'] = [0 if not i else \
                int(i[0][6]) if i[0][7] == ')' else int(i[0][6] + i[0][7])\
                for i in elevator]

cian_df['elevator_service'] = [0 if not i else \
                    int(i[1][7]) if len(i) == 2 else \
                    int(i[0][6]) if i and i[0][0] == 'Г' else 0\
                    for i in elevator]

cian_df['elevator_pass'] = cian_df['elevator_pass'].apply(lambda x: 1 if x <= 2 else 0)
cian_df['elevator_service'] = cian_df['elevator_service'].apply(lambda x: 1 if x < 1 else 0)

cian_df['floor_lower_20'] = cian_df['elevator_pass']*cian_df['elevator_service']
cian_df.drop(columns=['elevator_pass', 'elevator_service', 'Лифт'], inplace=True)
```

```python
#Категориальный призак высота потолков
cian_df.loc[cian_df['Высота потолков, м']>10,['Высота потолков, м']] = cian_df.loc[cian_df['Высота потолков, м']>10,['Высота потолков, м']]/100 
cian_df['ceiling_higher_3'] = cian_df['Высота потолков, м'] >= 3
cian_df['ceiling_higher_3'] = cian_df['ceiling_higher_3'].astype('int')
cian_df = cian_df.drop(columns='Высота потолков, м')
```
